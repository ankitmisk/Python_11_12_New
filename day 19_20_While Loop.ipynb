```python
# common elements

a = [1,2,3,45,5,6,76,78]
b = [4,5,6,7,8,89]

common = []

c = 0
while c<len(a):
    d  = 0
    while d<len(b):
        if a[c] == b[d]:
            common.append(a[c])
        d+=1
    c+=1
    
print(common)
```

```python
a = [1,2,3,45,5,6,76,78]
b = [4,5,3,6,7,8,89]
c = [4,5,3,6,7,8,90,0]

common = []



counter1 = 0
while counter1<len(a):
    
    counter2 = 0
    while counter2<len(b):
        
        counter3 = 0
        while counter3<len(c):
            
            if a[counter1] == b[counter2] == c[counter3]:
                
                common.append(a[counter1])
            
            counter3+=1
        
        counter2+=1
    
    counter1+=1
    
print(common)
```

```python
import random
import string


name_random = ['Karan','Ajay','lucky','mohan','ravi']
age_random = [random.randint(10,25) for i in range(5)]
class_random = [random.randint(5,12) for i in range(5)]
sec_random = [random.choice(string.ascii_lowercase) for i in range(5)]
add_random = ['Noida','Delhi','ranchi','Goa','Banglore']




data1 = {'Name':'Ajay','Age':23,'Class':12}
data2 = {'Sec':'A','Address':'Noida'}


no_of_data = 100

for m in data1:
    data1[m] = [data1[m]]
for n in data2:
    data2[n] = [data2[n]]
    
    
for i in range(no_of_data):
    for j in data1:
        ans1 = input(f'Enter your {j}: ').title()
        if j == 'Name':
            if ans1 == '':
                data1[j].append(random.choice(name_random).title())
            else:
                data1[j].append(ans1)
                
        elif j == 'Age':
            if ans1 == '':
                data1[j].append(random.choice(age_random))
            else:
                data1[j].append(ans1)
        
        elif j == 'Class':
            if ans1 == '':
                data1[j].append(random.choice(class_random))
            else:
                data1[j].append(ans1)
                
        else:
            pass
        
    for k in data2:
        ans2 = input(f'Enter your {k}: ')
        
        if k == 'Sec':
            if ans2 == '':
                data2[k].append(random.choice(sec_random).title())
            else:
                data2[k].append(ans2)
                
        elif k == 'Address':
            if ans2 == '':
                data2[k].append(random.choice(add_random).title())
            else:
                data2[k].append(ans2)
        
# print(data1)
# print(data2)

data1.update(data2)
import pandas as pd

table = pd.DataFrame(data1)

table
```
