# Encapsulation:

- Public variable
- protectec variable
- private variable

```python
class ATM:
    def __init__(self):
        self.pin = '0000'
        self.balance = 12345
        
    def withdrawl(self):
        print('Withdrawl Func!!')
    
    def deposit(self):
        print('Deposit func!!')
        
    def check_balance(self):
        print('This is a balance check func')
        print('Balance is',self.balance)
```

```python
hdfc = ATM()
```

```python
hdfc.check_balance()
```

```python
hdfc.balance
```

```python
hdfc.pin = '1234'
```

```python
hdfc.pin
```

```python
class ATM_Protected:
    def __init__(self):
        self._pin = '0000'
        self._balance = 12345
        
    def withdrawl(self):
        print('Withdrawl Func!!')
    
    def deposit(self):
        print('Deposit func!!')
        
    def check_balance(self):
        print('This is a balance check func')
        print('Balance is',self.balance)
```

```python
hdfc = ATM_Protected()
```

```python
# hdfc.pin = 1234
```

```python
hdfc.withdrawl()
```

```python
hdfc.deposit()
```

```python
hdfc._balance
```

```python
hdfc._pin = '123445'
```

```python
hdfc._pin
```

```python
class ATM_Private:
    def __init__(self):
        self.__pin = '0000'
        self.__balance = 12345
        
    def withdrawl(self):
        print('Withdrawl Func!!')
    
    def deposit(self):
        print('Deposit func!!')
        
    def check_balance(self):
        print('This is a balance check func')
        print('Balance is',self.balance)
```

```python
icic = ATM_Private()
```

```python
icic._ATM_Private__balance
```

```python
icic._ATM_Private__pin = '2345t6'
```

```python
icic._ATM_Private__pin
```

```python
icic.deposit()
```

```python
class ATM_Admin:
    def __init__(self):
        self.__pin = '0000'
        self.__balance = 12345
        
    def withdrawl(self):
        print('Withdrawl Func!!')
    
    def deposit(self):
        print('Deposit func!!')
        
    def check_balance(self):
        print('This is a balance check func')
        print('Balance is',self.__balance)
        
    def __adim(self):
        self.check_balance()
        self.deposit()
        self.withdrawl()
        print('All test PAss!!')
```

```python
axis = ATM_Admin()
```

```python
axis._ATM_Admin__adim()
```

#### multiple

```python
class AA:
    def a(self):
        print('I am a')
        
        
class BB:
    def b(self):
        print('I am b')
        
        
        
class CC:
    def c(self):
        print('I am c')
        
        
        
class DD:
    def d(self):
        print('I am d')
```

```python
class DD(AA,BB,CC):
    def d(self):
        print('I am d')
```

```python
obj_DD = DD()
```

```python
obj_DD.d()
```

```python
obj_DD.a()
```

```python
obj_DD.c()
```

```python
class Lava:
    def charger(self):
        print('This is a superfast charger!!')
        
class Sony:
    def camera(self):
        print('HUge Mega pixel camera!!')

class Samsung:
    def display(self):
        print('OLED Display High Quality!!')
```

```python
class Magma(Lava,Sony,Samsung):
    def phone_case(self):
        print('This is a phone case!!')
```

```python
sunny = Magma()
```

```python
sunny.phone_case()
```

```python
sunny.camera()
```

```python
sunny.display()
```

```python
sunny.charger()
```
