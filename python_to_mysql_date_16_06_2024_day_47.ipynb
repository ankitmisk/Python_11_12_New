```python
def give_data(limit):
    q = f'''select region,avg(surfacearea) as Average_surface_area
    from world.country
    group by region
    order by Average_surface_area desc
    limit {limit};'''

    return run_query(q)
```

```python
import random
```

```python
# sid	student_name	address	phone	class	sec	age
```

```python
def give_random_data(sid):
    student_name = 'Ajay karan ravi mohan shayam nikhil madan akshay sahil kumar aadil'.split(' ')
    address_val = 'noida delhi faridabad greaternoida lucknow gurugram'.split(' ')
    phone = random.randint(9876543210,9999999999)
    classs = random.randint(5,12)
    sec = list('ABCDEF')
    age = random.randint(15,23)

    final_data = []

    for i in range(1):
        final_data.append(sid)
        final_data.append(random.choice(student_name))
        final_data.append(random.choice(address_val))
        final_data.append(phone)
        final_data.append(classs)
        final_data.append(random.choice(sec))
        final_data.append(age)

    final_data = tuple(final_data)
    
    return final_data
```

```python
give_random_data(5)
```
